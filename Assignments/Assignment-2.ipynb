```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentDataSet.csv")
df.head()
```

```python
df.isnull()
```

```python
# Checkinf missing value using notnull()
series = pd.notnull(df["Reading_Score"])
print(series.tail())
```

```python
# Filling missing values using dropna(), fillna(), replace()

new_df = df.dropna()
print(df.tail())
```

```python
ndf = df.fillna(0)
ndf.tail()
```

```python
df["Placement_Score"] = df["Placement_Score"].fillna(df["Placement_Score"].mean())
df["Placement_Score"].head()
```

```python
plt.boxplot([df['Math_Score'],
             df['Reading_Score'],
             df['Writing_Score']])
plt.title("Student Score Distribution")
plt.show()
```

```python
print("Skewness of Math Score:", df["Math_Score"].skew())
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.exponential(scale=2, size=1000)

plt.hist(data, bins=30)
plt.title("Positive Skew (Right Skewed Distribution)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
